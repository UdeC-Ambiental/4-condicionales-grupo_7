```python
#carga de datos 
import pandas as pd
import numpy as np

df = pd.read_csv('muestras_calidad_agua.csv')
df.head()
```

```python
df_temp = df[['sitio', 'temperatura']]
df_temp.head()

#Extraer datos especificos 
```

```python
#datos de riesgo termico y categoria de temperatura 
df_temp['riesgo_termico'] = np.where(df_temp['temperatura'] > 28, 'Sí', 'No')
df_temp['categoria_temp'] = np.where(df_temp['temperatura'] < 20, 'Fría',
np.where(df_temp['temperatura'] <= 28, 'Moderada', 'Caliente'))

df_temp
```

```python
df_caliente = df_temp[df_temp['riesgo_termico'] == 'Sí']
df_caliente
```

```python
porcentaje_riesgo = (df_temp['riesgo_termico'].value_counts(normalize=True)['Sí']) * 100
print(f"Porcentaje de sitios con riesgo térmico: {porcentaje_riesgo:.2f}%")
```

```python
#linea para guardar la tabla final de "resultado_temperatura" en ruta de archivo csv
df_temp.to_csv('resultado_temperatura.csv', index=False)
```
